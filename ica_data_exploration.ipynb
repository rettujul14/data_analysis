```python
#From the above box we have identified outliers for the age field and for the HNR field we are planning to run log transformation to make the values to be match with other features.
#Remove outlier for age
q1=df['age'].quantile(0.25)
q3=df['age'].quantile(0.75)
iqr=q3-q1
#Interquartile range
df_out = df
df_out = df.loc[(df['age'] > (q1-1.5*iqr)) & (df['age'] < (q3+1.5*iqr))]

plt.boxplot(df_out["age"])
plt.show()
```

```python
#Run log transformation for HNR column
df_out["HNR"] = df_out["HNR"].map(lambda i: np.log(i) if i > 0 else 0) 
plt.boxplot(df_out["HNR"])
plt.show()
```

```python
#Heatmap to simpy identify
import seaborn as sns

plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
#From the above heat map we can see lot of columns with correlation. So am setting the correlation threshold to 0.9
def correlation(dataset,threshold):
    col_corr = set() # set all the names of correlated columns
    corr_matrix = dataset.corr() # hey just execute  dataset.corr() to understand this line.
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if(corr_matrix.iloc[i,j])> threshold:
                colname= corr_matrix.columns[i]  #getting column names
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(df_out,0.9)
len(set(corr_features))
corr_features
```
